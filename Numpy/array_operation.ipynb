```python
#stacking array
```

```python
a=np.array([1,2,3,4])
b=np.array([4,5,6,7])
```

```python
np.vstack((a,b))  #vertical add arrays
```

```python
np.hstack((a,b))  #horizontal add arrays
```

```python
np.column_stack((a,b))   #1row->1column ans 2rows->2column
```

```python
c = np.arange(1,37).reshape(6,6)
c
```

```python
np.hsplit(c,3)   #  note -np.hsplit in NumPy is used to split an array horizontally (column-wise) into multiple sub-arrays.
#only split in proper no of cut like 6 is cut by 1,2, 3 or 6 
```

```python
np.vsplit(c,3)    #  note -np.hsplit in NumPy is used to split an array vertically (rows-wise) into multiple sub-arrays. 
#only split in proper no of cut like 6 is cut by 1,2, 3 or 6 #(if its a odd no then we whato cut with 2.5)
```

```python
for i in c:
    print(i)
```
