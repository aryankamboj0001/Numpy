```python
import seaborn as sns
import plotly.express as px

tips = sns.load_dataset('tips')

fig = px.line(tips, y='total_bill', title='Total Bill Trend')
fig.show()
```

```python


mean_tips = tips.groupby('day', as_index=False)['tip'].mean()

fig = px.bar(mean_tips, x='day', y='tip', title='Average Tip by Day')
fig.show()
```

```python
fig = px.scatter(tips, x='total_bill', y='tip', color='day',
                 title='Total Bill vs Tip (by Day)')
fig.show()
```

```python
fig = px.box(tips, x='day', y='total_bill',
             title='Total Bill Distribution by Day')
fig.show()
```
