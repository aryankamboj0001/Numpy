1) which team win most number of matches?

```python
match_win = df["match_winner"].value_counts()
```

```python
sns.barplot(x=match_win.values,y=match_win.index,palette="rainbow")
plt.title("most win by the team")
```

2) Toss Decision Trends

```python
sns.countplot(x=df['toss_decision'],palette="rainbow")
plt.title('toss decision trends')
```

3) Toss Winner vs Match Winner 

```python
count = df[df['toss_winner']==df['match_winner']]['match_id'].count()
count
```

```python
percentage = (count*100)/df.shape[0]
percentage
```

4) How do teams win? (Runs vs Wickets)

```python
sns.countplot(x=df['won_by'])
plt.title("won by")
```

# Key Player Performances

1) Most "Player of the Match" Awards

```python
count = df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(y=count.index,x=count.values,palette='rainbow')
plt.title('player of the match')
```

```python
df.head()
```

2) Top Scorers

```python
high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='bar')   #if we put h after bar then it will become horizontal
```

```python
high.plot(kind='barh')
```

3) ten Best Bowling Figures

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x : x.split('--')[0])
df['highest_wickets'] = df['highest_wickets'].astype(int)
top_bowler = df.groupby('best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowler
```

```python
top_bowler.plot(kind='barh')
```

# Venue Analysis
