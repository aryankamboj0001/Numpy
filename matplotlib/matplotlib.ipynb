Object Oriented Way

Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches dpi is the dots-per-inch (pixel per inch).

```python
fig = plt.figure(figsize=(16,9),dpi=100)  #figsize means aspect ratio , dpi means quality by defoult it is 100
axis1 = fig.add_axes([0.1,0.1,1,1])  #varies from 0-1 and 0 means o percent and 100 means 100 percent and like .3 = 30% , .5 = 50%
axis1.plot(x,y)
axis2 = fig.add_axes([.3,.4,.2,.5])  #.3= move left on x asis , .4 means move above y axis , .2 means 20% size on x axis , .5 means 50%size on y axis
axis2.plot(x,y)
```
