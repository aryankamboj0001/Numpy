```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset('tips')
df
```

```python
df['size'].unique()
```

```python
plt.subplot(1,2,1)
sns.histplot(df['total_bill'],bins=50,kde=True)  #default bins = 20  , #instead of histplot you can also use displot both work same
plt.subplot(1,2,2)
sns.histplot(df['tip'],bins=20,kde=True)
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='kde')
```

```python
sns.pairplot(df,hue='sex',palette="rainbow")
```
