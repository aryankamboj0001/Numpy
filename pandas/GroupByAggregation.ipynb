Group By

```python
data = {
    'category' : ['A','A','B','A','B','A','B','B'],
    'store' : ['S1','S2','S1','S2','S1','S1','S2','S2'],
    'sales' : [100,200,115,150,200,500,700,450],
    'date' : pd.date_range('2023-01-01',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

```python
cat = df.groupby('category')
for i,v in cat:
    print(i)
    print(v)
```

```python
cat = df.groupby('category')['sales'].sum()
cat
```

```python
cat = df.groupby("store")['sales'].sum()
cat
```

```python
cat = df.groupby(['category','store'])['sales'].sum()
cat
```

```python
cat = df.groupby(['store','category'])['sales'].sum()
cat
```
