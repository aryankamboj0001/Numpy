```python
import numpy as np
import pandas as pd
```

```python
#finding missing data
```

```python
data={
    'A' : [1,np.nan,3,4,5],
    'B' : [1,2,3,4,5],
    'C' : [1 ,np.nan,3,np.nan,5],
    'D' : [np.nan,2,3,np.nan,np.nan]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any()
```

```python
#removing missing data
```

```python
df
```

```python
df.dropna()   #it work rows wise    only give non null rows
```

```python
df.dropna(thresh=3)   #gives the rows in which the element in rows has 3 or more non-null element
```

```python
#filling the missing data
```

```python

df
```

```python
df.fillna(0)  #fill the non-null value
```

```python
values={'A':0,'B':100,'C':200,'D':400}   #it work column wise
df.fillna(value=values)
```

```python
df.fillna(df.mean())
```
