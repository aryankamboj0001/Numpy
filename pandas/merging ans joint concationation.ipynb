```python
import numpy as np
import pandas as pd
```

```python
#merging 2 dataframe
```

```python
employees = pd.DataFrame({
    'employee_ID':[1,2,3,4,5],
    'name' :['Aryan','Kamboj','Virat','Kohli','Rohit'],
    'department' : ['HR','IT','Finance','IT','HR']
})

salaries = pd.DataFrame({
    'employee_ID':[1,2,3,6,7],
    'salart': [60000,50000,45000,70000,20000],
    'bonus':[1000,3000,500,1000,300]
    
})
```

```python
employees
```

```python
salaries
```

```python
pd.merge(employees,salaries,on='employee_ID',how='inner')  #on means common column and how means only non null rows
```

```python
pd.merge(employees,salaries,on='employee_ID',how='outer')    #outer give all null and non noll rows
```

```python
pd.merge(employees,salaries,on='employee_ID',how='left')  #on the basis of left means element put in merge like in this employees
```

```python
pd.merge(employees,salaries,on='employee_ID',how='right')   #on the basis of right means element put in merge like in this salaries
```

**Concationation of 2 dataframes**

```python
df1 = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B1','B2','B3'],
    'C': ['C1','C2','C3']
})
df2 = pd.DataFrame({
    'A': ['A3','A4','A5'],
    'B': ['B3','B4','B5'],
    'C': ['C3','C4','C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df1,df2])
```

```python
pd.concat([df1,df2],axis = 1)
```

```python
pd.concat([df1,df2],axis = 0)
```
